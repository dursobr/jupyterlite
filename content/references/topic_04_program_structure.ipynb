## Functions

Functions help you:
- **Avoid repetition** (write code once, use it many times)
- **Organize** long programs into readable pieces
- **Test** and debug smaller parts more easily
- **Communicate intent** (good names + docstrings)

Compare repeated code vs. a function.

````python
import math
import matplotlib.pyplot as plt
````

````python
# Repeated code (not ideal)
g = 9.81
h1 = 2.0
h2 = 10.0
h3 = 25.0

v1 = math.sqrt(2*g*h1)
v2 = math.sqrt(2*g*h2)
v3 = math.sqrt(2*g*h3)

print(v1, v2, v3)
````

Now the same idea with a function.

In physics, the speed of an object dropped from height \(h\) (starting from rest, no air resistance) is:

$v = \sqrt{2gh}$

````python
def drop_speed_from_height(h, g=9.81):
    return math.sqrt(2*g*h)

for h in [2.0, 10.0, 25.0]:
    print(h, drop_speed_from_height(h))
````

### Anatomy of a function

A function definition has:
- `def` keyword
- A **name** (use descriptive names)
- A parameter list `(x, y, ...)`
- An indented body
- Optional `return` statement

General form:
```python
def name(parameters):
    '''Docstring'''
    # body
    return result
```

### Parameters vs. arguments
- **Parameters**: variables in the function definition
- **Arguments**: actual values you pass when calling the function

````python
def kinetic_energy(m, v):
    """Compute kinetic energy.

    Parameters
    ----------
    m : float
        Mass in kilograms.
    v : float
        Speed in meters per second.

    Returns
    -------
    float
        Kinetic energy in joules.
    """
    return 0.5*m*v**2

print(kinetic_energy(0.145, 40))  # a baseball at 40 m/s
````

### Return values (and multiple returns)

A function can return:
- one value (most common)
- multiple values (as a tuple)

Physics example: constant-acceleration kinematics.

$y(t) = y_0 + v_0 t + \tfrac{1}{2} a t^2$

$v(t) = v_0 + a t$

````python
def kinematics_1d(t, y0, v0, a):
    """Position and velocity for 1D constant acceleration.

    Parameters
    ----------
    t : float
        Time in seconds.
    y0 : float
        Initial position in meters.
    v0 : float
        Initial velocity in meters per second.
    a : float
        Constant acceleration in m/s^2.

    Returns
    -------
    (float, float)
        (y, v) position in meters and velocity in m/s.
    """
    y = y0 + v0*t + 0.5*a*t**2
    v = v0 + a*t
    return y, v

print(kinematics_1d(2.0, y0=10.0, v0=0.0, a=-9.81))
````

### Default parameters & keyword arguments

Defaults make functions easier to use.
- Default parameters go at the end: `def f(x, g=9.81): ...`
- Keyword arguments let you call by name: `f(h=10, g=9.8)`

Physics example: projectile motion (no air resistance).

For launch speed $v$ and angle $theta$, the **range** (flat ground) is:
$R = \frac{v^2 \sin(2\theta)}{g}$

````python
def projectile_range(v, theta_deg, g=9.81):
    """Compute horizontal range for ideal projectile motion (flat ground).

    Parameters
    ----------
    v : float
        Launch speed in m/s.
    theta_deg : float
        Launch angle in degrees.
    g : float, optional
        Gravitational acceleration (m/s^2), default 9.81.

    Returns
    -------
    float
        Range in meters.
    """
    theta = math.radians(theta_deg)
    return (v**2 * math.sin(2*theta)) / g

print(projectile_range(20, 45))
print(projectile_range(v=20, theta_deg=30, g=9.8))
````

## Docstrings: structure and best practices

A **docstring** is the triple-quoted string right after the `def` line.

Good docstrings usually include:
- **One-line summary** (what the function does)
- **Longer description** (optional)
- **Parameters** (names, types, units, meaning)
- **Returns** (what comes back)
- **Raises** (errors/exceptions)
- **Examples** (how to use it)

There are multiple style conventions.
In this notebook we use a *NumPy/SciPy-style* layout because it is common in science.

## Exceptions: raising and catching errors

In Python, **exceptions** are how the language signals that something went wrong.

Common examples:
- `ZeroDivisionError` — dividing by zero
- `ValueError` — a value is the right type but not acceptable (e.g., negative mass)
- `TypeError` — wrong type (e.g., string where a number is expected)

Why exceptions matter in science/physics code:
- They help you **fail early** when inputs are not physically meaningful.
- They make bugs easier to find.
- They let you handle problems gracefully (e.g., skip bad data and keep going).

### Basic idea
- **Raise** an exception when your function cannot proceed.
- **Catch** an exception with `try/except` when you want to recover.

### Example: Python raises exceptions automatically

````python
# Uncomment to see the exception
# 1 / 0

# Another common one:
# int('not a number')
````

### Raising exceptions in your own functions

When you validate inputs (units, ranges, physical constraints), raise an appropriate exception.

Physics example: period of a simple pendulum (small-angle approximation)

$T = 2\pi\sqrt{\frac{L}{g}}$

Physical constraints:
- Length $L$ must be **positive**
- Gravity $g$ must be **positive**

````python
def pendulum_period(L, g=9.81):
    """Compute the period of a simple pendulum (small-angle approximation).

    Parameters
    ----------
    L : float
        Pendulum length in meters (must be > 0).
    g : float, optional
        Gravitational acceleration magnitude in m/s^2 (must be > 0).

    Returns
    -------
    float
        Period in seconds.

    Raises
    ------
    ValueError
        If L <= 0 or g <= 0.
    """
    if L <= 0:
        raise ValueError("L must be > 0 (pendulum length must be positive)")
    if g <= 0:
        raise ValueError("g must be > 0")
    return 2*math.pi*math.sqrt(L/g)

print(pendulum_period(1.0))
# print(pendulum_period(-1.0))  # try this
````

### Catching exceptions with `try/except`

Use `try/except` when you expect something *might* fail and you want to continue.

Example: computing periods for a list of lengths, some of which may be invalid.

````python
lengths = [1.0, 0.5, -0.2, 2.0, 0.0]
periods = []

for L in lengths:
    try:
        T = pendulum_period(L)
        periods.append(T)
    except ValueError as e:
        print(f"Skipping L={L}: {e}")

print("Valid periods:", periods)
````

### Catching multiple exceptions and using `else` / `finally`

- Catch multiple exceptions by providing a tuple: `except (TypeError, ValueError) as e:`
- `else` runs **only if no exception occurred**
- `finally` runs **no matter what** (useful for cleanup)

Physics example: computing wave speed on a string:
$v = \sqrt{\frac{T}{\mu}}$

where tension $T>0$ and linear mass density $mu>0$.

````python
def wave_speed_string(tension, mu):
    """Compute wave speed on a string: v = sqrt(T/mu).

    Parameters
    ----------
    tension : float
        Tension in newtons (must be > 0).
    mu : float
        Linear mass density in kg/m (must be > 0).

    Returns
    -------
    float
        Wave speed in m/s.

    Raises
    ------
    ValueError
        If tension <= 0 or mu <= 0.
    """
    if tension <= 0 or mu <= 0:
        raise ValueError("tension and mu must both be > 0")
    return math.sqrt(tension/mu)

cases = [(100, 0.01), (50, 0), ("hi", 0.02), (80, 0.02)]

for tension, mu in cases:
    try:
        v = wave_speed_string(tension, mu)
    except (TypeError, ValueError) as e:
        print(f"Bad inputs {tension=}, {mu=}: {e}")
    else:
        print(f"OK: {tension=}, {mu=} -> v={v:.2f} m/s")
    finally:
        # This always runs (often used to close files or release resources)
        pass
````

### Designing your own exception messages

Good exception messages:
- say *what* was wrong
- include the bad value when helpful
- suggest the valid range/units

Example: improving `drop_speed_from_height` error message.

````python
def drop_speed_from_height_checked(h, g=9.81):
    """Return impact speed for an object dropped from rest, with input checks.

    Parameters
    ----------
    h : float
        Drop height (m), must be >= 0.
    g : float, optional
        Gravitational acceleration (m/s^2), must be > 0.

    Returns
    -------
    float
        Impact speed (m/s).

    Raises
    ------
    ValueError
        If h < 0 or g <= 0.
    """
    if g <= 0:
        raise ValueError(f"g must be > 0 m/s^2 (got g={g})")
    if h < 0:
        raise ValueError(f"h must be >= 0 m (got h={h})")
    return math.sqrt(2*g*h)

for h in [10, -3]:
    try:
        print(h, drop_speed_from_height_checked(h))
    except ValueError as e:
        print("Error:", e)
````

## Functions that validate inputs

Defensive programming: check for physically meaningful inputs.

Physics example: Newton's law of gravitation
$F = G\frac{m_1 m_2}{r^2}$

````python
def gravitational_force(m1, m2, r, G=6.67430e-11):
    """Compute the magnitude of gravitational force between two masses.

    Parameters
    ----------
    m1, m2 : float
        Masses in kilograms.
    r : float
        Separation distance in meters (must be > 0).
    G : float, optional
        Gravitational constant (N·m^2/kg^2).

    Returns
    -------
    float
        Force magnitude in newtons.

    Raises
    ------
    ValueError
        If r <= 0 or if any mass is negative.

    Examples
    --------
    Force between Earth and a 1 kg mass at Earth's surface (approx.):
    >>> gravitational_force(5.972e24, 1.0, 6.371e6)
    9.8...
    """
    if r <= 0:
        raise ValueError("r must be positive")
    if m1 < 0 or m2 < 0:
        raise ValueError("masses must be non-negative")
    return G*m1*m2/(r**2)

print(gravitational_force(5.972e24, 1.0, 6.371e6))
````
