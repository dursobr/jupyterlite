## Exercise: Projectile Motion
The vertical position of a projectile is given by

$y = y_0 + v_0 t - \frac{1}{2}g t^2$

where $y_0$ is the initial vertical position, $v_0$ is the initial vertical speed and $g = 9.8$ m/s$^2$.

Make a plot of the vertical position of a projectile over time with $y_0 = 0$ m and $v_0 = 12.3$ m/s from launch until it returns to the initial height. Label the axes.

````python
import numpy as np
import matplotlib.pyplot as plt

g = 9.8 # m/s^2
v_0 = 12.3 # m/s
t = np.linspace(0.0, 2.51, 100)
y = v_0*t - 0.5*g*t**2

plt.plot(t, y)
plt.xlabel("Time (s)")
plt.ylabel("Vertical Position (m)")
````

## Simple `if` Statements
A simple `if` statement runs code **only if** a condition is true.

There is no alternative action if the condition is false.

````python
temperature = 120  # degrees Celsius

if temperature > 100:
    print("Warning: Water is boiling!")
````

If the condition is false, Python simply skips the indented block.

## Why Indentation Matters
Python uses **indentation** to decide which lines of code belong together.

Indented code runs *only* when the condition above it is true.

````python
# Correct indentation
x = 5

if x > 0:
    print("x is positive")
    print("This line is also conditional")

print("This line always runs")
````

If indentation is wrong, Python will raise an error or behave incorrectly.

**Rule of thumb:**
- All lines in the same block must be indented the same amount
- Standard practice is **4 spaces**

## `if / else` Statements
An `if / else` statement allows Python to choose **between two paths**.

Exactly one of the blocks will run.

````python
velocity = 3  # m/s

if velocity == 0:
    print("The object is at rest.")
else:
    print("The object is moving.")
````

## `if / elif / else` Statements
`elif` (else-if) allows more than two choices.

This is common when classifying physical behavior.

````python
velocity = 15  # m/s

if velocity == 0:
    print("The object is at rest.")
elif velocity < 10:
    print("The object is moving slowly.")
else:
    print("The object is moving fast.")
````

## `while` Loops
**Key idea:** A `while` loop repeats **as long as** a condition is `True`.

### Syntax
```python
while condition:
    # code that repeats
```

- Python checks `condition` **before** each repetition.
- If it is `True`, the indented block runs.
- If it is `False`, the loop stops.

### When to use `while`
Use `while` when you **don't know in advance** how many repetitions you need or when it is inconvenient to make many copies of code to do something repeatedly.

Examples:
- Keep asking for input until the user enters a valid value
- Keep simulating motion until an object hits the ground
- Keep iterating until the error is below a tolerance


### Example: counting up
Let's count from 1 to 5 using a `while` loop.

````python
count = 1
while count <= 5:
    print(count)
    count += 1  # IMPORTANT: update the variable so the loop can end
````

### Avoiding infinite loops
A loop becomes **infinite** if the condition never becomes `False`.

Common causes:
- Forgetting to update a variable used in the condition
- Updating it in the wrong direction
- Condition is always `True`

> **Tip:** If you accidentally run an infinite loop in Jupyter, click **Kernel → Interrupt**.


### `break` and `continue`

- `break` **exits** the loop immediately.
- `continue` **skips** the rest of the current repetition and goes back to the condition.

Example: find the first power of 2 that exceeds 1000.


````python
n = 0
while True:  # intentionally 'infinite' until we break
    value = 2**n
    if value > 1000:
        print(f"2^{n} = {value} is the first power of 2 greater than 1000")
        break
    n += 1
````

## Exercise: Calculating Prime Numbers
Use nested `while` loops to find and print the first `N_requested` prime numbers.

````python
N_requested = 10 # number of primes requested
n = 2  # first potential prime
 
# outer loop: Keep going until we have N_requested primes
N = 0
while N < N_requested:
    is_prime = True
    divisor = 2
    
    # inner loop: check if n is divisible by any number from 2 up to n-1
    while divisor < n:
        if n % divisor == 0:
            is_prime = False
            break
        divisor += 1
        
    if is_prime:
        print(n)
        N += 1
        
    n += 1 # move to the next number to check
````

## Lists
A **list** stores multiple values in a single variable. The values do not need to have the same `type`.

In physics, lists often represent:
- Measurements over time
- Positions, velocities, or energies
- Experimental data

````python
# Positions of a particle measured every second (meters)
positions = [0.0, 1.2, 2.8, 4.5, 6.1]
print(positions)
````

### List Creation
A list is created by placing items (elements) inside square brackets `[]`, separated by commas.

````python
# Creating a list of strings
fruits = ['apple', 'banana', 'cherry']

# Lists can hold different data types
mixed_list = [10, 'hello', 3.14, True]

print("Fruits list:", fruits)
print("Mixed list:", mixed_list)
````

### Indexing and Slicing
Python uses **zero-based indexing**, meaning the first item is at index `0`.

````python
numbers = [10, 20, 30, 40, 50, 60]

print("First item (index 0):", numbers[0])
print("Last item (index -1):", numbers[-1])

# Slicing: [start:stop] - 'stop' index is exclusive
print("Items from index 1 to 3:", numbers[1:4])

# Slicing with step: [start:stop:step]
print("Every second item:", numbers[::2])
````

### List Manipulation
Lists are **mutable**, meaning you can change their content after they are created.

````python
colors = ['red', 'blue', 'green']

# Change an item
colors[1] = 'yellow'

# Add an item to the end
colors.append('purple')

# Insert an item at a specific position
colors.insert(1, 'orange')

# Remove an item by value
colors.remove('red')

print("Updated colors:", colors)
````

### Useful List Methods
Python provides built-in methods to perform common operations.

````python
my_list = [5, 2, 9, 1, 5, 6]

print("List length:", len(my_list))
print("Count of '5':", my_list.count(5))

my_list.sort()
print("Sorted list:", my_list)

my_list.reverse()
print("Reversed list:", my_list)
````

## Example: The Arithmetic Sequence
An athlete runs distance $d_1 = 5$ km on Day 1. Every day after, they increase their distance by $1.5$ km. Use a `for` loop and the formula below to calculate the distance on Day $n$ and list the distances for the first week. Plot the distances versus day with red dots and plot a black line on top of the distances to show the arithmetic sequence is linear. Label the axes.

**Formula:** $d_n = d_{n-1} + 1.5$

````python
import matplotlib.pyplot as plt

d = 5 # km
distances = []
days = list(range(1, 8))

for n in days:
    distances.append(d)
    d += 1.5 # or d = d + 1.5

print("Days:", days)
print("Distances:", distances, "km")

plt.plot(days, distances, marker='o', linestyle='', color='red')
plt.plot([days[0], days[-1]], [distances[0], distances[-1]], color='black')
plt.title('Arithmetic Sequence: Linear Growth')
plt.xlabel('Time (day)')
plt.ylabel('Distance (km)')
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
````

## The `range()` function
The `range()` function is used to generate a sequence of numbers. It is most commonly used in `for` loops to repeat an action a specific number of times.

### The Basic Range: `range(stop)`
If you provide one argument, it starts at **0** and stops **before** that number.

````python
print("Range(5):")
for i in range(5):
    print(i)  # Output: 0, 1, 2, 3, 4
````

### Setting a Start: `range(start, stop)`
You can specify where the sequence begins. Note that the 'stop' value is **exclusive**.

````python
print("Range(2, 7):")
for i in range(2, 7):
    print(i)  # Output: 2, 3, 4, 5, 6
````

### Adding a Step: `range(start, stop, step)`
The 'step' determines the increment between numbers. You can even use a negative step to count backwards!

````python
print("Counting by 2s:")
print(list(range(0, 11, 2))) 

print("\nCounting Backwards:")
print(list(range(10, 0, -1)))
````

### Why use `list()`?
By itself, `range()` is a generator object (it creates numbers on the fly to save memory). To see the full sequence as a list, we wrap it in `list()`.

````python
r = range(5)
print("Raw range object:", r)
print("Converted to list:", list(r))
````

## Example

````python
# Compute displacements between successive positions
positions = [0.0, 1.2, 2.8, 4.5, 6.1]
displacements = []

for i in range(len(positions)-1):
    dx = positions[i+1] - positions[i]
    displacements.append(dx)

displacements
````

## Plotting Motion
We can directly plot lists with `matplotlib`!

````python
import matplotlib.pyplot as plt

t = list(range(len(positions)))

plt.figure()
plt.plot(t, positions)
plt.xlabel("Time (s)")
plt.ylabel("Position (m)")
plt.title("Position vs Time")
plt.show()
````

## Example: Average Speed

````python
dt = 1.0  # seconds between measurements

total_distance = 0.0

for dx in displacements:
    if dx < 0:
        print("Moving backwards!")
    total_distance += abs(dx)

average_speed = total_distance / (dt * len(displacements))
average_speed
````
